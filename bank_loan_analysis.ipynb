```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("loan_data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
print(df.columns)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
total_applications = len(df)
print("Total Applications:", total_applications)
```

```python
avg_interest = df['int.rate'].mean()
print("Average Interest Rate:", avg_interest)
```

```python
avg_fico = df['fico'].mean()
print("Average FICO Score:", avg_fico)
```

```python
total_installment = df['installment'].sum()
print("Total Installment:", total_installment)
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='purpose', data=df)
plt.xticks(rotation=45)
plt.title("Loan Purpose Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x='credit.policy', data=df)
plt.title("Credit Policy Status")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x='not.fully.paid', data=df)
plt.title("Loan Repayment Status")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['fico'], bins=30)
plt.title("FICO Score Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='fico', y='installment', data=df)
plt.title("FICO Score vs Installment")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['int.rate'], bins=30)
plt.title("Interest Rate Distribution")
plt.show()
```

```python
plt.figure(figsize=(12,8))

numeric_df = df.select_dtypes(include=['number'])

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.show()
```

```python
df.to_csv("cleaned_loan_data.csv", index=False)
```
